*   actual_unit_Aug = 1000 (actual units sold in August)

2. Calculate sales revenue achieved in August
*   actual_rev_Aug = rev_unit * actual_unit_Aug

3. Calculate number of units to be sold in remaining months (Sep to Dec) to meet the sales revenue target for the year
*   expected_rev_Sep_Dec = target_rev_annual - actual_rev_Jan_Jul - actual_rev_Aug
*   expected_unit_Sep_Dec = expected_rev_Sep_Dec / rev_unit

4. print the values for sales revenue for August and units to be sold in Sep to Dec to achieve target




```python
# As per details given
rev_unit = 22.2
target_rev_annual = 800000
actual_rev_Jan_Jul = 335220
actual_unit_Aug = 1000

#Calculate sales revenue achived in August
actual_rev_Aug = rev_unit * actual_unit_Aug

#Calculate expected revenue and units to meet annual target
expected_rev_Sep_Dec = target_rev_annual - (actual_rev_Jan_Jul + actual_rev_Aug)
expected_unit_Sep_Dec = expected_rev_Sep_Dec / rev_unit

#Print the results
print("Sales revenue achieved for August = ", actual_rev_Aug)
print("Units expected to be sold during Sep to Dec to achive annual target = ", expected_unit_Sep_Dec)
```
