**MODEL01**

```python
# Import necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Import necessary libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler

# Load training and test data
train = pd.read_csv(r"/content/Santander Customer Satisfaction - TRAIN.csv")
test = pd.read_csv(r"/content/Santander Customer Satisfaction - TEST-Without TARGET.csv")

# Separate features and target variable from the training data
X = train.drop(columns=['ID', 'TARGET'])  # 'ID' is identifier, 'TARGET' is the target variable
y = train['TARGET']

# Prepare test features
X_test = test.drop(columns=['ID'])

# Optional: Standardize features (helps with model performance)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(X_test)

# Initialize the Decision Tree Classifier
dt_model = DecisionTreeClassifier(criterion='entropy', max_depth=150, random_state=1, splitter='random')
param_grid = {
    'min_samples_split': [100, 500, 1000, 3000]
}
param_grid = {
    'min_samples_leaf': [100, 350, 700, 1000]
}

# Train the model
dt_model.fit(X_scaled, y)

# Predict probabilities on the test data
y_test_prob = dt_model.predict_proba(X_test_scaled)[:, 1]  # Probability of class TARGET=1

# Apply threshold to determine final predictions
y_test_pred = [1 if prob > 0.5 else 0 for prob in y_test_prob]

# Create submission DataFrame with 'ID' and 'TARGET' columns
submission = pd.DataFrame({
    'ID': test['ID'],
    'TARGET': y_test_pred
})

# Save submission file
submission.to_csv("CIS508_GarimaAsths_MODEL01.csv", index=False)
print("Submission file saved as CIS508_GarimaAsths_MODEL01.csv")
```
